```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
df = pd.read_excel("D:/Data Science for Marketing-I/data/Online Retail.xlsx")
```

```python
df.info()
```

```python
df=df.loc[df['Quantity']>0]
```

```python
df['CustomerID'].isna().sum()
```

```python
df=df.dropna(subset=['CustomerID'])
```

```python
df.shape
```

```python

customer_item_matrix = df.pivot_table(index='CustomerID', columns='StockCode', values='Quantity', aggfunc='sum')
```

```python
customer_item_matrix.loc[12451:].head()
```

```python
customer_item_matrix.loc[12481.0].sum()
```

```python
customer_item_matrix=customer_item_matrix.applymap(lambda x: 1 if x>0 else 0)
```

```python
customer_item_matrix.loc[12481.0].sum()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = pd.DataFrame( cosine_similarity(customer_item_matrix))
```

```python
user_similarity.head()
```

```python
user_similarity.columns = customer_item_matrix.index
user_similarity.index = customer_item_matrix.index
user_similarity.head()
```
