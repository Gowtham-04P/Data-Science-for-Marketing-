### v.	How can conversion rates by the number of contacts be calculated and visualized in a dataset

```python
conversion_by_contacts = df.groupby('campaign')['conversion'].mean()
conversion_by_contacts.plot(kind='line', marker='o', color='cyan', markersize=5, linewidth=2)
plt.title("Conversion Rate by Number of Contacts")
plt.xlabel("Number of Contacts ")
plt.ylabel("Conversion Rate(%)")
plt.grid(True)
plt.show()
```

 üí° 
The data shows a positive correlation between the number of contacts and conversion rates. While increasing contacts improves conversions, it‚Äôs important to find the optimal number of contacts to balance effort and results.Beyond a certain point (e.g., 50 contacts), the conversion rate may not increase significantly, so further efforts might not be cost-effective.

### vi.	How are age groups created using a lambda function in a dataset, and why is grouping data into age ranges beneficial for analysis?

```python
df['age_group']=df['age'].apply(lambda x:'-30' if x<=30 else '31-40' if x<=40 else '41-50' if x<=50 else '51-60' if x<=60 else '60+')
df['age_group'].value_counts()
```

```python
df.groupby('age_group')['conversion'].sum()/df.groupby('age_group')['conversion'].count()*100
```

 üí°The age group 31-40 has the highest count (16,385 individuals), making it the largest group.The age group 41-50 is the second-largest, with 10,240 individuals.The age group -30 (likely representing ages 0-30) has 7,383 individuals.The age group 51-60 has 6,270 individuals.The age group 60+ has the smallest count, with only 910 individuals.
 Grouping data into age ranges is beneficial for analysis because it helps identify trends, patterns, or behaviors within specific age groups, simplifying insights and decision-making.

### vii.	In an analysis comparing conversions and non-conversions by marital status, what additional insights could be explored and how would you extend the code to perform this analysis with the variable Education 

```python
marital_status_conversion = df.groupby(['marital', 'conversion']).size().unstack()
education_conversion = df.groupby(['education', 'conversion']).size().unstack()
marital_status_conversion.plot(kind='bar', stacked=True, figsize=(8, 5), color=['gold', 'red'])
plt.title("Conversions by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.legend(["No Conversion", "conversion"])
plt.show()

```

```python
education_conversion.plot(kind='bar', stacked=True, figsize=(8, 5), color=['red', 'gold'])
plt.title("Conversions by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.legend(["No Conversion", "conversion"])
plt.show()
```

```python
# Group by marital status and education, and calculate conversion rates
conversion_analysis = df.groupby(['marital', 'education'])['conversion'].agg(['sum', 'count'])
conversion_analysis['conversion_rate'] = conversion_analysis['sum'] / conversion_analysis['count'] * 100
conversion_analysis = conversion_analysis.reset_index()

# Display the results
print(conversion_analysis)
```

```python
df.groupby(['marital','education'])['conversion'].mean().unstack()
```

```python
import seaborn as sns
```

```python
# Create a pivot table for conversion rates
pivot_table = conversion_analysis.pivot(index='marital', columns='education', values='conversion_rate')

# Plot a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Conversion Rate (%)'})
plt.title('Conversion Rates by Marital Status and Education Level')
plt.xlabel('Education Level')
plt.ylabel('Marital Status')
plt.show()
```
